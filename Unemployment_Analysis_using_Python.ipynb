```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/content/Unemployment in India.csv')
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Check data types 
df.info()
```

```python
#check the Summary Statistics
df.describe()

```

```python
#Check for missing values
df.isnull().sum()
```

```python
# print the column names 
df.columns
```

```python
# Visualize the distribution of Estimated Unemployment Rate
sns.histplot(df[' Estimated Unemployment Rate (%)'], bins=10)
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Estimated Unemployment Rate')
plt.show()
```

```python
# Visualize the relationship between Estimated Employed and Estimated Labour Participation Rate
sns.scatterplot(data=df, x=' Estimated Employed', y=' Estimated Labour Participation Rate (%)')
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Labour Participation Rate (%)')
plt.title('Relationship between Estimated Employed and Estimated Labour Participation Rate')
plt.show()
```

```python
# Calculate correlation between variables
corr_matrix = df.corr()
print(corr_matrix)
```

```python
# Group data by Region and calculate average Estimated Unemployment Rate
avg_unemployment_rate = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
print(avg_unemployment_rate)
```

```python
# Create a bar chart of average Estimated Unemployment Rate by Region
sns.barplot(data=df, x='Region', y=' Estimated Unemployment Rate (%)')
plt.xlabel('Region')
plt.ylabel('Average Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.title('Average Estimated Unemployment Rate by Region')
plt.show()
```
